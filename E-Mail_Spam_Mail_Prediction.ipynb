Importing the Dependencies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

Data Collection & Pre-Processing

```python
raw_mail_data = pd.read_csv(r'C:\Users\Hp\OneDrive\Desktop\EI System\mail_data.csv')
```

```python
print(raw_mail_data)
```

```python
mail_data = raw_mail_data.where((pd.notnull(raw_mail_data)),'')
```

```python
mail_data.shape
```

EDA

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=mail_data, x='Category')
plt.title('Spam vs Ham Distribution')
plt.show()
```

```python
mail_data['Message Length'] = mail_data['Message'].apply(len)
plt.figure(figsize=(12, 6))
sns.histplot(data=mail_data, x='Message Length', hue='Category', bins=50, kde=True)
plt.title('Distribution of Message Lengths')
plt.show()
```

Label Encoding

```python
mail_data.loc[mail_data['Category'] == 'spam', 'Category',] = 0
mail_data.loc[mail_data['Category'] == 'ham', 'Category',] = 1
```

```python
X = mail_data['Message']

Y = mail_data['Category']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & test data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=3)
```

```python
print(X.shape)
print(X_train.shape)
print(X_test.shape)
```

Feature Extraction

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words='english', lowercase=True)

X_train_features = feature_extraction.fit_transform(X_train)
X_test_features = feature_extraction.transform(X_test)


Y_train = Y_train.astype('int')
Y_test = Y_test.astype('int')
```

```python
print(X_train)
```

```python
print(X_train_features)
```

Training the Model

Logistic Regression

```python
model = LogisticRegression()
```

```python
model.fit(X_train_features, Y_train)
```

```python
def plot_top_tfidf_words(class_labels, n_words=20):
    feature_names = np.array(feature_extraction.get_feature_names_out())
    for label in np.unique(class_labels):
        label_name = 'Spam' if label == 0 else 'Ham'
        top_n_idx = np.argsort(model.coef_[0])[-n_words:] if label == 1 else np.argsort(model.coef_[0])[:n_words]
        top_n_words = feature_names[top_n_idx]
        top_n_weights = model.coef_[0][top_n_idx]

        plt.figure(figsize=(10, 5))
        sns.barplot(x=top_n_weights, y=top_n_words)
        plt.title(f'Top {n_words} TF-IDF Features for {label_name}')
        plt.show()

plot_top_tfidf_words(Y_train)
```

Evaluating the trained model

```python
prediction_on_training_data = model.predict(X_train_features)
accuracy_on_training_data = accuracy_score(Y_train, prediction_on_training_data)
```

```python
print('Accuracy on training data : ', accuracy_on_training_data)
```

```python
prediction_on_test_data = model.predict(X_test_features)
accuracy_on_test_data = accuracy_score(Y_test, prediction_on_test_data)
```

```python
print('Accuracy on test data : ', accuracy_on_test_data)
```

```python
conf_matrix = confusion_matrix(Y_test, prediction_on_test_data)
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.show()
```

```python
class_report = classification_report(Y_test, prediction_on_test_data, target_names=['Spam', 'Ham'])
print('Classification Report:\n', class_report)
```

Building a Predictive System

```python
def predict_mail(input_mail):
    input_data_features = feature_extraction.transform(input_mail)
    prediction = model.predict(input_data_features)
    
    if prediction[0] == 1:
        return 'Ham mail'
    else:
        return 'Spam mail'

while True:
    user_input = input("Enter a message to classify (or type 'exit' to quit): ")
    if user_input.lower() == 'exit':
        break
    prediction = predict_mail([user_input])
    print(f"The message is classified as: {prediction}")
```
